```python
import numpy as np
import matplotlib.pyplot as plt

# Создание данных
array_3d = np.zeros((18, 18, 18))
 # Это будут ваши значения
for i in range(18):
    for j in range(18):
        for k in range(18):
            array_3d[i][j][k] = np.random.rand(1)
# Создание массива цветов, например, на основе значений z
x, y, z = np.indices(array_3d.shape)
from scipy.ndimage.filters import gaussian_filter

colors = array_3d

# Создание трехмерного графика
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Построение точек с цветами
sc = ax.scatter(x,y,z, c=colors, cmap='plasma', s=10)

# Добавление цветной шкалы
cbar = plt.colorbar(sc)
cbar.set_label('Значения')

plt.show()
```
