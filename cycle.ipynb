```python
import matplotlib.pyplot as plt
import numpy as np

#   (  )
data = {
    str(i): np.random.randint(1, 50) for i in range(24)
}

#   ()   (),    
hours = [i for i in reversed(range(24))]
values = list(data.values())

#    ,  0   
angles = np.linspace(0, 2 * np.pi, len(hours), endpoint=False).tolist()
angles = angles[7:] + angles[:7]  #    

#    
fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

#  ,      0  6 
ax.plot([*angles,angles[0]], [*values,values[0]], linewidth=2, linestyle='-', color='skyblue')
ax.fill_between(angles[17:], 0, values[17:], color='green', alpha=0.5)  #   

#   ()
ax.set_xticks(angles)
ax.set_xticklabels(hours)

#  
plt.title('   ')

#  
plt.show()
```
